```python
cmorph = pd.read_csv('/content/drive/MyDrive/Grupo 4 - TEEP/Trabalho Prático - Parte II/df_cmorph.csv')
alertario = pd.read_csv('/content/drive/MyDrive/Grupo 4 - TEEP/Trabalho Prático - Parte II/krig_UTC_chuva_horaria_alertario.csv')
```

```python
indice_cmorph = cmorph['DataHora'].values
alerta_filtrado = alertario.loc[alertario['DataHora'].isin(indice_cmorph)]
```

```python
alerta_filtrado['DataHora'] = pd.to_datetime(alerta_filtrado['DataHora'], format='%Y-%m-%d %H:%M:%S')
datas_extremas = [pd.date_range('1998-02-11 01:00:00', freq = 'h', periods = 24),
                  pd.date_range('2001-12-23 01:00:00', freq = 'h', periods = 24),
                  pd.date_range('2003-03-17 01:00:00', freq = 'h', periods = 24),
                  pd.date_range('2007-10-24 01:00:00', freq = 'h', periods = 24),
                  pd.date_range('2010-04-05 01:00:00', freq = 'h', periods = 24),
                  pd.date_range('2019-04-08 01:00:00', freq = 'h', periods = 24),
                  pd.date_range('2022-04-30 01:00:00', freq = 'h', periods = 24)]
# Transforma a lista em uma lista que pode ser comparada com o dataset
datas_extremas = [item for sublist in datas_extremas for item in sublist]
# erro = erro.reset_index(names = 'DataHora')
alerta_eventos_extremos = alerta_filtrado.loc[alerta_filtrado["DataHora"].isin(datas_extremas)] # Começar a partir das 01:00 horas
alerta_eventos_extremos.to_csv('/content/drive/MyDrive/Grupo 4 - TEEP/Trabalho Prático - Parte II/df_alerta_eventos_extremos.csv', index = False)
alerta_eventos_extremos
```

```python
alerta_filtrado = alerta_filtrado.select_dtypes(include=['float64']).reset_index(drop=True)
```

```python
cmorph_filtrado = cmorph.loc[:,cmorph.columns.isin(alertario.columns)]
cmorph_filtrado['DataHora'] = pd.to_datetime(cmorph_filtrado['DataHora'], format='%Y-%m-%d %H:%M:%S')
datas_extremas = [pd.date_range('1998-02-11 01:00:00', freq = 'h', periods = 24),
                  pd.date_range('2001-12-23 01:00:00', freq = 'h', periods = 24),
                  pd.date_range('2003-03-17 01:00:00', freq = 'h', periods = 24),
                  pd.date_range('2007-10-24 01:00:00', freq = 'h', periods = 24),
                  pd.date_range('2010-04-05 01:00:00', freq = 'h', periods = 24),
                  pd.date_range('2019-04-08 01:00:00', freq = 'h', periods = 24),
                  pd.date_range('2022-04-30 01:00:00', freq = 'h', periods = 24)]
# Transforma a lista em uma lista que pode ser comparada com o dataset
datas_extremas = [item for sublist in datas_extremas for item in sublist]
# erro = erro.reset_index(names = 'DataHora')
cmorph_eventos_extremos = cmorph_filtrado.loc[cmorph_filtrado["DataHora"].isin(datas_extremas)] # Começar a partir das 01:00 horas
cmorph_eventos_extremos.to_csv('/content/drive/MyDrive/Grupo 4 - TEEP/Trabalho Prático - Parte II/df_cmorph_eventos_extremos.csv', index = False)
cmorph_eventos_extremos
```
